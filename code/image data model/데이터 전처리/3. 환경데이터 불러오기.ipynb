```python
# 필요한 라이브러리 불러오기
import numpy as np
import pandas as pd

import json
import os
```

```python
# 빈 데이터프레임 생성
apple_df = pd.DataFrame()
val_df = pd.DataFrame()
```

## json 파일 모두 합치기

```python
# json파일 데이터프레임으로 통합(훈련용)
path1 = '/content/gdrive/MyDrive/리미트리스(limitless)/핵심역량프로젝트/데이터/data/3차/train'
file_list1 = os.listdir(path1)
file_list_py1 = [file for file in file_list1 if file.endswith('.json')]
file_list_len1 = len(file_list_py1)
dict_list1 = []
for i in file_list_py1:
    try:
        for line in open((path1 + '/' + i), "r", encoding='utf8'):
            dict_list1.append(json.loads(line))
    except:
        pass
        
    df1 = pd.DataFrame(dict_list1)
    df1 = pd.concat([apple_df, df1], ignore_index = True)
```

```python
# json파일 데이터프레임으로 통합(검증용)
path2 = '/content/gdrive/MyDrive/리미트리스(limitless)/핵심역량프로젝트/데이터/data/3차/val'
file_list2 = os.listdir(path2)
file_list_py2 = [file for file in file_list2 if file.endswith('.json')]
file_list_len2 = len(file_list_py2)
dict_list2 = []
for i in file_list_py2:
    try:
        for line in open((path2 + '/' + i), "r", encoding='utf8'):
            dict_list2.append(json.loads(line))
    except:
        pass
        
    df2 = pd.DataFrame(dict_list2)
    df2 = pd.concat([val_df, df2], ignore_index = True)
```

```python
#훈련용 데이터 확인
df1
```

## 딕셔너리 꼴의 데이터를 이용하기 편하게 바꾸기

```python
# 필요부분만을 활용한 데이터프레임 생성(딕셔너리 꼴의 데이터에서 필요한 key값을 열로 지정 후 짝인 value값 넣기)

for col in df1.columns:

    val_df1 = pd.DataFrame()

    for i in range(len(df1.index)):
        col_dict1 = df1[col][i]
        val_dict1 = {}

        for keys in col_dict1.keys():
            if type(col_dict1.get(keys)) is list:
                val_dict1[keys] = [col_dict1.get(keys)]
            else:
                df1.loc[i, keys] = col_dict1.get(keys)
        
        # list가 value값인 열이 맨 마지막에 있었기 때문에 가능한 것 (val_df는 새로운 열에서 초기화 됨)
        val_dict1 = pd.DataFrame(val_dict1, index = [i])
        val_df1 = pd.concat([val_df1, val_dict1])

    df1 = df1.drop(col, axis = 1)
```

```python
# 리스트가 값인 열
val_df1.head(5)
```

```python
df1 = pd.concat([df1, val_df1], axis = 1)
df1
```

```python
# 딕셔너리 꼴의 데이터에서 필요한 key값을 열로 지정 후 짝인 value값 넣기

for col in df2.columns:

    val_df2 = pd.DataFrame()

    for i in range(len(df2.index)):
        col_dict2 = df2[col][i]
        val_dict2 = {}

        for keys in col_dict2.keys():
            if type(col_dict2.get(keys)) is list:
                val_dict2[keys] = [col_dict2.get(keys)]
            else:
                df2.loc[i, keys] = col_dict2.get(keys)
        
        # list가 value값인 열이 맨 마지막에 있었기 때문에 가능한 것 (val_df는 새로운 열에서 초기화 됨)
        val_dict2 = pd.DataFrame(val_dict2, index = [i])
        val_df2 = pd.concat([val_df2, val_dict2])

    df2 = df2.drop(col, axis = 1)
```

```python
# 리스트가 값인 열
val_df2.head(5)
```

```python
df2 = pd.concat([df2, val_df2], axis = 1)
df2
```

## 필요없는 열 삭제

```python
df1 = df1.drop(['description','url','version','year','img_path','label_path'], axis = 1)
df2 = df2.drop(['description','url','version','year','img_path','label_path'], axis = 1)
```
