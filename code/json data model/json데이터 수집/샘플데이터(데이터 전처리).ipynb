```python
## 해당 경로에 있는 .json 파일명 리스트 가져오기

import os

path = '/content/drive/MyDrive/리미트리스(limitless)/핵심역량프로젝트/데이터/sample'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.json')] ## 파일명 끝이 .json인 경우
```

```python
## json 파일들을 DataFrame으로 불러오기

import json

dict_list = []
for i in file_list_py:
    for line in open((path+'/'+i),"r"):
        dict_list.append(json.loads(line))
df = pd.DataFrame(dict_list)
```

```python
df
```

```python
# 필요없는 부분 삭제(info = 기본정보(데이터셋 이름, 제작자,제작버전,제작년도,데이터셋 타입, 이미지데이터 폴더경로, 라벨링 데이터 폴더 경로), licenses = 저작권정보(라이센스 고유번호, 라이센스 이름))
df = df.drop(['info','licenses'], axis = 1) 
```

```python
df1 = df['collection'] # 수집정보 데이터프레임
df2 = df['images'] # 이미지 데이터정보 데이터프레임
df3 = df['annotations'] # 어노테이션정보(segmentation 좌표, 어노테이션 면적, bounding box 정보, 사과 당도 품질 클래스(A,B,C)) 데이터프레임
```

```python
# 데이터 프레임 형태로 변환
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = pd.DataFrame(df3)
```

```python
# for문에 사용할 빈 리스트, 딕셔너리 생성
df1_ = {}
df1_list = []
df2_ = {}
df2_list = []
df3_1_ = {}
df3_1_list = []
df3_2_ = {}
df3_2_list = []
```

```python
# 샘플데이터 전체 df1(수집정보 데이터프레임) 생성
for i in range(100):
    df1_[i] = df1['collection'][i]
    df1_[i] = pd.DataFrame.from_dict(df1_[i], orient = 'index')
    df1_list.append(df1_[i])
    apple_df1 = pd.concat(df1_list, axis =1, ignore_index=True)
```

```python
# 샘플데이터 전체 df2(이미지 데이터정보 데이터프레임) 생성
for i in range(100):
    df2_[i] = df2['images'][i]
    df2_[i] = pd.DataFrame.from_dict(df2_[i], orient = 'index')
    df2_list.append(df2_[i])
    apple_df2 = pd.concat(df2_list, axis =1, ignore_index=True)
```

```python
# 샘플데이터 df3_1(사과 당도 품질 클래스 데이터프레임) 생성
for i in range(100):
    df3_1_[i] = df3['annotations'][i]
    df3_1_[i] = df3_1_[i]['sugar_grade']
    df3_1_list.append(df3_1_[i])
    apple_df3_1 = pd.DataFrame(df3_1_list)
```

```python
# 샘플데이터 df3_2(segmentation 데이터프레임) 생성
for i in range(100):
    df3_2_[i] = df3['annotations'][i]
    df3_2_[i] = df3_2_[i]['segmentation']
    df3_2_list.append(df3_2_[i])
    apple_df3_2 = pd.DataFrame(df3_2_list)
```

```python
#행,열 변환
apple_df1 = apple_df1.transpose() 
apple_df2 = apple_df2.transpose()
```
